```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('files/combine.csv',delim_whitespace=True,header=0,skipfooter=1,skipinitialspace=True,names=['Time','d','uwall.bias','uwall.force2','metad.bias','metad.work','ct'])
data['ct'] = data['ct'].str.replace('"', '')

data.index = range(len(data))
data.tail()
data.head()
```

```python
#names=['Time','d','uwall.bias','uwall.force2','metad.bias','metad.work','ct']
z=[]
m=[]
for i in range(len(data)):
    
    x=data['metad.bias'].iloc[i]
    y=float(data['ct'].iloc[i])
    z.append(np.exp(x-y))
    m.append(np.exp(x-y+20)/100000000)

data['e(v-ct)']=z
data['scaled_e']=m
data.head(50)
#data.tail(50)
```

```python
number=np.sum(data['scaled_e'])
print(number)
no=[]
for i in range(len(data)):
    no.append(data['scaled_e'].iloc[i]/number)
data['Normalized']=no
data.head(100)
```
